
# Dimensionality Reduction 
Dimensionality reduction is a technique used in machine learning to reduce the number of input variables (features) in a dataset while retaining as much relevant information as possible. This helps improve computational efficiency, reduce overfitting, and enhance visualization.

# Why is Dimensionality Reduction Important?
1. <b>Reduces Overfitting</b>:Too many features can lead to models that perform well on training data but poorly on new data (high variance).
2. <b>Improves Model Performance</b>: Reducing unnecessary features can speed up  training and inference and reduces computational cost.
3. <b>Better Visualization</b>: Data with high dimensions (e.g., 100+ features) is hard to interpret. Reducing it to 2D or 3D helps in visualization.


# Types of Dimensionality Reduction Techniques
1. Feature Selection
2. Feature Extraction

# Feature Selection
Feature selection is the process of choosing the most relevant features (input variables) for a machine learning model. This helps improve performance, reduce overfitting, and speed up computations. Selects a subset of relevant features from the original dataset.

# Feature Extraction
Feature extraction is a dimensionality reduction technique that transforms raw data into a set of new, informative features. Unlike feature selection, which removes irrelevant features, feature extraction creates new features from the original ones.

# PCA
Principal Component Analysis (PCA) is a linear dimensionality reduction technique that transforms high-dimensional data into a lower-dimensional space while retaining as much variance as possible. It does so by finding new axes (principal components) along which the data varies the most. PAC is a unsupervised learning.

variance is propogetion of spread

# Advantages & Disadvantages of PCA
✅ <b>Advantages</b>
* ✔ Reduces computation time and storage.
* ✔ Removes multicollinearity among features.
* ✔ Helps in data visualization.

❌ <b>Disadvantages</b>
* ❌ PCA assumes linear relationships and may not work well with non-linear data.
* ❌ The transformed components lose interpretability (e.g., "PC1" and "PC2" instead of meaningful features).
* ❌ PCA doesn’t work well when data is not standardized.

# Grid Search
Grid Search is a hyperparameter tuning technique used to find the best combination of hyperparameters for a machine learning model. It systematically evaluates all possible combinations of hyperparameters and selects the one that results in the best model performance.
